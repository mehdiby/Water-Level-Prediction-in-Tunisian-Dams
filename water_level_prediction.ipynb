# Benchmarking of Algorithms

```python
# Define the list of targets and the feature columns
targets = ['MELLEGUE', 'BEN METIR', 'KASSEB', 'BARBARA', 'SIDI SALEM',
       'BOU-HEURTMA', 'JOUMINE', 'GHEZALA', 'SEJNANE', 'S. EL BARRAK',
       'SILIANA', 'LAKHMESS', 'RMIL', "BIR M'CHERGA", 'RMEL', 'NEBHANA',
       'SIDI SAAD', 'EL HAOUAREB', 'SIDI AÏCH', 'EL BREK', 'BEZIRK', 'CHIBA',
       'MASRI', 'LEBNA', 'HMA', 'ABID', 'Zarga', 'Ziatine']
features = ['prcp', 'dayofweek', 'month', 'quarter', 'year']

# Playing with features after studying Correlation 
#features = ['tavg', 'tmin', 'tmax', 'prcp', 'snow', 'wdir', 'wspd', 'wpgt', 'pres', 'tsun', 'dayofweek', 'month', 'quarter', 'year']
```

```python
df_test_mock = df_test

# Loop over the targets
for target in targets:
    # Split the data into a training set and a test set
    merged_df_1 = merged_df[[target, *features]]
    merged_df_1 = merged_df_1.dropna()

    # assuming the data is stored in a pandas DataFrame called 'df'
    profile = pp.ProfileReport(merged_df_1)

    # Save the report to an HTML file
    # profile.to_file(f"report_{target}.html")

    # profile.to_widgets()

    X_train, X_test, y_train, y_test = train_test_split(merged_df_1[features], merged_df_1[target], test_size=0.2)
    models = [
        LinearRegression(),
        RandomForestRegressor(n_estimators=50000),
        XGBRegressor(n_estimators=50000, learning_rate=0.001, max_depth=8),
        GradientBoostingRegressor(n_estimators=50000, max_depth=8),
        DecisionTreeRegressor(max_depth=8),
        SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1),
        ElasticNet(alpha=0.01, l1_ratio=0.5),
        Lasso(alpha=0.01),
        Ridge(alpha=0.05),
    ]
    best_model = models[0]
    best_rmse = 1000
    # Create an instance
    for model in models:
      print(model)
      # Train the model on the training set
      model.fit(X_train, y_train)

      # Use the model to make predictions on the test set
      y_pred = model.predict(X_test)

      # Evaluate the performance of the model using RMSE
      rmse = np.sqrt(mean_squared_error(y_test, y_pred))
      print(f"RMSE for {target}: {rmse}")
      if rmse < best_rmse:
        best_rmse = rmse
        best_model = model

    # Use the model to make predictions on the test set
    df_test_mock[target] = best_model.predict(merged_df_test[features])
    #print(f"Best model for {target}: {type(best_model).__name__}")
```

```python
# Loop over the targets
for target in targets:
  # Try lgb
  # Set the hyperparameters
  params = {
      'objective': 'regression',
      'metric': 'rmse',
      'boosting_type': 'gbdt'
  }

  # Train the model
  train_data = lgb.Dataset(X_train, label=y_train)
  model = lgb.train(params, train_data, num_boost_round=10000)
  # Make predictions
  y_pred = model.predict(X_test)
  rmse = np.sqrt(mean_squared_error(y_test, y_pred))
  print(f"LGB RMSE for {target}: {rmse}")
```

```python
# Bring best models and try to have the mean form them.
for target in targets:
    # Split the data into a training set and a test set
    merged_df_1 = merged_df[[target, *features]]
    merged_df_1 = merged_df_1.dropna()

    X_train, X_test, y_train, y_test = train_test_split(merged_df_1[features], merged_df_1[target], test_size=0.2)

    df_test_mock[target] = 0
    models = [
        RandomForestRegressor(n_estimators=50000),
        XGBRegressor(n_estimators=50000, learning_rate=0.001, max_depth=8),
        GradientBoostingRegressor(n_estimators=50000, max_depth=8),
    ]

    # Create an instance
    for model in models:
      print(model)
      # Train the model on the training set
      model.fit(X_train, y_train)

      # Use the model to make predictions on the test set
      y_pred = model.predict(X_test)

      # Evaluate the performance of the model using RMSE
      rmse = np.sqrt(mean_squared_error(y_test, y_pred))
      print(f"RMSE for {target}: {rmse}")
      df_test_mock[target] += model.predict(merged_df_test[features])
      
    # Use the model to make predictions on the test set
    df_test_mock[target]/3
```
